```python
from pprint import pprint
from math import log
import numpy as np
import pandas as pd
import operator
```

```python
rb=[]
jd=[]
text1 = open('nath.txt',encoding='utf-8-sig')
for x in text1:
    rb.append(x.strip())
for b in rb:
    if b=='':
        rb.remove(b)
#print(rb[4],"\n",len(rb),"\n",rb)



text2 = open('jibon.txt',encoding='utf-8-sig')
for x in text2:
    jd.append(x.strip())
for b in jd:
    if b=='':
        jd.remove(b)
#print(jd[4],"\n",len(jd),"\n",jd)
```

```python
data=rb.copy()
data.extend(jd)
print(data)
```

```python
def mysplt(a):
    return a.split(' ')
```

```python
def wordCalculate(data):
    uniquW={}
    for x in data:
        for y in mysplt(x):
            if y !='':
                if y in uniquW:
                   uniquW[y]+=1
                else:
                   uniquW[y]=1
    uniquW=sorted(uniquW.items(), key=lambda x: x[1], reverse=True)
    return dict(uniquW)
```

```python
alldata=wordCalculate(data)
print(alldata,len(alldata))
```

```python
nath=wordCalculate(rb)
print(nath,"\n\n",len(nath))
```

```python
jibon=wordCalculate(jd)
print(jibon,"\n\n",len(jibon))
```

```python
total_data_len=len(data)
total_nath_len=len(nath)
total_jibon_len=len(jibon)
alldata_len=len(alldata)
all_prob_nath=total_nath_len/total_data_len
all_prob_jibon=total_jibon_len/total_data_len
print(all_prob_jibon," ",all_prob_nath)
```

```python
prob_word={}
for word,wkey in alldata.items():
    pb=wkey/alldata_len
    prob_word[word]=pb
print(prob_word)
```

```python
silgne_prob_nath={}
for word,wkey in alldata.items():
    wkey=1
    if word in nath:
        wkey+=nath[word]
    pb=wkey/(total_nath_len+2)
    silgne_prob_nath[word]=pb
print(silgne_prob_nath)
```

```python
single_prob_jibon={}
for word,wkey in alldata.items():
    wkey=1
    if word in jibon:
        wkey+=jibon[word]
    pb=wkey/(total_jibon_len+2)
    single_prob_jibon[word]=pb
print(single_prob_jibon)
```

```python
test=[]
read_test = open('test.txt',encoding='utf-8-sig')
for x in read_test:
    test.append(x.strip())
for b in test:
    if b=='':
        test.remove(b)

        

        
final_nath=1
for x in test:
    for word in mysplt(x):
        prob_in_nath=silgne_prob_nath[word] / prob_word[word]
        final_nath*=prob_in_nath
    final_nath*=prob_in_nath
#print(final_nath)


final_jibon=1
for x in test:
    for word in mysplt(x):
        prob_in_jibon=single_prob_jibon[word] / prob_word[word]
        final_jibon*=prob_in_jibon
final_jibon*=prob_in_jibon
#print(final_jibon)

if final_nath >= final_jibon:
    print('লেখক রবীন্দ্রনাথ ঠাকুর')
else:
    print('লেখক জীবনানন্দ দাশ')
```
