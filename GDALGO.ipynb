```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(12,10)

data=pd.read_csv('salary_data.csv')
X=data.iloc[:, 0]
Y=data.iloc[:, 1]
plt.scatter(X,Y)
plt.show()
```

```python
m=0
c=0
l=0.0001
myit=1000
n=float(len(X))
for i in range(myit):
    my=m*X+c
    dm=(-2/n)* sum(X*(Y-my))
    dc=(-2/n)* sum(Y-my)
    m=m-l*dm
    c=c-l*dc
print(m,c)
```

```python
my=m*X+c
plt.scatter(X,Y)
plt.plot(X,my,color="red")
plt.show()
```
